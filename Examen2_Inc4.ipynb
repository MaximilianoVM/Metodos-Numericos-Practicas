Importamos librerias

```python
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

definimos la funcion y su derivada

```python
def f(x): 
    weon = x*x*x*x + 2*x*x -x -3 
    return weon

def df(x):
    weon = 4*x*x*x + 4*x - 1
    return weon
```

```python
i=0
Po = 2
aproximaciones = []
while i< 3 :
    print("_______ciclo " + str(i+1))
    p = Po - ( f(Po) )/( df(Po) )

    print( "Po = " + str(Po) + "\n" + "P = " + str(p) )
    print( "p-Po = " + str( abs(p-Po) ))

    print( "aproximacion " + str(i+1) + "a la raíz: p= " + str(p) )

    if( abs(p-Po) < 1e-5 ): break

    aproximaciones.append(p)

    i += 1
    Po=p

print("======================= RESUMEN =======================")
print("las aproximaciones son: ")
for i in range(len(aproximaciones)): 
    print("p"+ str(i+1) + "= " + str(aproximaciones[i]))
    print("f(p)=" + str(f(p)))
```

```python
for i in range(len(aproximaciones)): 

    x = np.arange(-2,3,0.1)
    y = f(x)
    plt.ylim(-4,10)

    plt.plot(x,y, linestyle='-',linewidth=3)
    plt.grid(True,linewidth=3)

    plt.plot(aproximaciones[i] ,f(aproximaciones[i]), '-o',linewidth=5)

    #recta
    plt.plot(x, df(aproximaciones[i]) * (x-aproximaciones[i]) + (f(aproximaciones[i])) , linestyle='-',linewidth=3) 

    plt.xlabel('$x$',fontsize=20)
    plt.ylabel('$f(x)$',fontsize=20)

    plt.title('Aproximacion con p'+str(i+1)+"= "+str(aproximaciones[i]),fontsize=20)

    plt.show()
```
