Definimos las funciones

```python
from cmath import sqrt


def f1(x): 
    weon = np.tan(np.pi * x) - 6
    return weon
def f2(x): 
    weon = np.cos(x) - x
    return weon
def f3(x):
    weon =  np.sin(np.sqrt(x)) - x
    return weon

##DERIVADAS

def df1(x):
    weon = np.pi * (1 / np.cos(np.pi * x) )**2
    return weon
def df2(x):
    weon = - np.sin(x) - 1
    return weon
def df3(x):
    weon = ( np.cos( sqrt(x) ) / (2 * sqrt(x)) )
    return weon
    
```

<h1>/////////-------Metodo de la biseccion-------/////////</h1>

definimos arreglos necesarios para desplegar despues las tablas

```python
arreglof1 = []
arreglof2 = []
arreglof3 = []
```

Definimos el metodo

```python
def bisec(a, b, n, TOL, NF): 
    i = 1

    ###fA###
    if NF==1: 
        fA = f1(a)
    elif NF==2: 
        fA = f2(a)
    elif NF==3: 
        fA = f3(a)
    ########

    while i<=n :

        p = a + (b-a)/2

        ###fP###
        if NF==1: 
            fP = f1(p)
            arreglof1.append([i, a, b, p, fP])
        elif NF==2: 
            fP = f2(p)
            arreglof2.append([i, a, b, p, fP])
        elif NF==3: 
            fP = f3(p)
            arreglof3.append([i, a, b, p, fP])
        ########

        #if fP==0 or abs(p - a)<=TOL: 
        #    print("SE CUMPLIO MENOR A TOL")
        #    print("numero de iteraciones: " + str(i))
        #    print("solucion aproximada: " + "\n" + "p' = " + str(p))
        #    break

        if abs(fP) <= TOL and abs( p - Po ): 
            print("SE CUMPLIO EL CRITERIO DE PARO")
            print("numero de iteraciones: " + str(i))
            print("solucion aproximada: " + "\n" + "p' = " + str(p))
            break

        #contador y Po := p_n-1, en realidad tambien nos sirve en este caso a = p_n-1
        i+=1
        Po = p

        if fA*fP > 0: 
            a = p
            fA = fP 
        else: 
            b = p
    print("===========!!!!!!!=============")
    print("el metodo concluyo despues de " + str(i) + " iteraciones")
    
```

<h1>Ejecucion de la funcion</h1><br> 
bisec(a, b, n, TOL, NF), <br>
donde: <br>
<b>n</b> es el numero maximo de iteraciones<br>
<b>NF</b> es el numero de la funcion a utilizar en el metodo (1 = f1 , 2 = f2 , 3 = f3).<br>

<h1>/////////-------Metodo de Newton Raphson-------/////////</h1>

```python
arreglof12 = []
arreglof22 = []
arreglof32 = []
```

Definimos el metodo
