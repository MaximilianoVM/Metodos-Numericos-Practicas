<h1>Practica 3: Metodo de iteracion de punto fijo</h1>

<p>A continuacion, se define el programa principal con el que computaremos los puntos fijos:</p>

```python
#Librerias necesarias para las funciones y las graficas en python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#Casos a evaluar
p0_1 = 1.1
p0_2 = 1.3

#funcion de la que queremos las raices
def f(x):
    return (x*x*x) + (4*x*x) -10

#Arreglos en donde seran almacenadas las filas de las tablas
arregloG1 = []
arregloG2 = []
arregloG3 = []
arregloG4 = []
arregloG5 = []

#Definimos las funciones de iteracion de punto fijo a utilizar
def g1(x):
    return x - (x*x*x) - (4*(x*x)) + 10
def g2(x):
    return ((10 / x) - 4*x) ** (1/2)
def g3(x):
    return (1 / 2)*(10 - x*x*x) ** (1/2)
def g4(x):
    return (10 / (4 + x)) ** (1/2)
def g5(x):
    return x - (((x*x*x) + (4*(x*x)) - 10) / ((3*(x*x)) + (8*x)))

#Variables para definir el total de iteraciones y la tolerancia para la condicion de parada
n = 30
tolerancia = 10**(-10)

#Funcion para determinar puntos fijos
def fixedPoint(p0, g):

    i = 1
    while i <= n:
        
        #Condicionales para determinar la funcion propuesta a utilizar. Una vez elegida la funcion, se evalua p0 en la misma y se agrega el resultado a uno de los arreglos.
        if g ==1 :
            p = g1(p0)
            arregloG1.append([i, p, g1(p)])
        elif g == 2 :
            p = g2(p0)
            arregloG2.append([i, p, g2(p)])
        elif g == 3 :
            p = g3(p0)
            arregloG3.append([i, p, g3(p)])
        elif g == 4 :
            p = g4(p0)
            arregloG4.append([i, p, g4(p)])
        elif g == 5 :
            p = g5(p0)
            arregloG5.append([i, p, g5(p)])
        else:
            return -1 #Retorna -1 cuando el g(x) en el argumento de la funcion no esta definido
        
        #Condicion de parada para el ciclo
        if abs(p - p0) < tolerancia:
            return p
        
        #Aumenta el indice y establece el nuevo valor de p0
        i = i + 1
        p0 = p
        
    return -1 #Si la funcion retorna -1, entonces no converge al punto fijo

fixedPoint(p0_1, 1)
fixedPoint(p0_1, 2)
fixedPoint(p0_1, 3)
fixedPoint(p0_1, 4)
fixedPoint(p0_1, 5)

tabla1 = np.array(arregloG1)
tabla2 = np.array(arregloG2)
tabla3 = np.array(arregloG3)
tabla4 = np.array(arregloG4)
tabla5 = np.array(arregloG5)
```

```python
#BUSQUEDA VISUAL DEL 0 
x=np.arange(1.34, 1.4, 0.01)
v=f(x)
fig=plt.figure()
plt.plot(x,v, linestyle='-',linewidth=3)
plt.xlabel('x',fontsize=20)
plt.ylabel('f(x)',fontsize=20)
plt.title('b√∫squeda visual del cero',fontsize=20)
plt.grid(True,linewidth=5)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```
